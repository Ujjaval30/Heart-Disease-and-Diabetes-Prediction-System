```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score
```

```python
heart_data=pd.read_csv("heart_disease_data.csv")
```

```python
X=heart_data.drop(columns="target",axis=1)
Y=heart_data.target
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
X.shape,X_train.shape,X_test.shape
```

```python
model=LogisticRegression(random_state=2,max_iter=1000)
```

```python
model.fit(X_train.values,Y_train.values)
```

```python
X_train_prediction=model.predict(X_train.values)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on training data is:",training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test.values)
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on testing data is:",testing_data_accuracy)
```

```python
input_data=(58,1,2,112,230,0,0,165,0,2.5,1,1,3)
input_data_as_array=np.asarray(input_data)
input_data_reshped=input_data_as_array.reshape(1,-1)
prediction=model.predict(input_data_reshped)
print(prediction)
if(prediction[0]==0):
    print("The patient is free from any heart disease")
else:
    print("Patient has heart disease")
```
